```python
import matplotlib.pyplot as plt
location = "./data/texts.txt"
with open(location, "r", encoding="utf-8") as file:
    text = file.read()
```

```python
from collections import Counter
import string
words = text.translate(str.maketrans("", "", string.punctuation)).lower().split()
freqs_with_outliers = Counter(words).values()
freqs = [x for x in freqs_with_outliers if x not in {}]
max_fr = max(freqs)
min_fr = min(freqs)

normalized_freqs = [(fr - min_fr)/(max_fr - min_fr) for fr in freqs]


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))
ax1.hist(freqs, 20)
ax1.set_title('frequencies histo')
ax2.hist(normalized_freqs, 20)
ax2.set_title('[0,1] scaled frequencis histo')
plt.show()
```
