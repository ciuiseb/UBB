```python
location = "./data/texts.txt"
with open(location, "r", encoding="utf-8") as file:
    text = file.read()
print(text)
```

```python
print(f"Numarul de propozitii este {text.count('.') + text.count('!') + text.count('?')}")
# gpt zice ca sunt 7 propozitii hmmm
```

```python
import string
from collections import Counter

words = text.translate(str.maketrans("", "", string.punctuation)).lower().split()

print(f"Numarul de cuvinte este {len(words)}")
print(f"Numarul de cuvinte diferite este {len(Counter(words))}")
print(f"Cel mai scurt cuvant este {min(words, key=len)}")
print(f"Cel mai lung cuvant este {max(words, key=len)}")
```

```python
diacritice = str.maketrans(
    'ăâîșțĂÂÎȘȚ',
    'aaistAAIST'
)
print(text.translate(diacritice))
```
